```python
#Print the columns names containing null values or missing values in the form of a list
print(athletes[athletes.Age.isnull()])
print(athletes[athletes.Height.isnull()])
print(athletes[athletes.Weight.isnull()])
print(athletes[athletes.Medal.isnull()])
```

```python
#Top countries participating
countries = athletes['Team'].value_counts().sort_values(ascending=False).head(10)
```

```python
countries
```

```python
#Plot for the top 10 countries
pt.figure(figsize=(12,6))
pt.title('Overall Participation by Country')
sns.barplot(x = countries.index, y=countries)
```

```python
#Using Matplotlib
pt.figure(figsize=(12,5))
ax =athletes['Team'].value_counts().sort_values(ascending=False).head(10).plot(kind='bar')
ax.set_xlabel('Countries')
ax.set_ylabel('Team')
ax.set_title('Overall participation of countries')
```

```python
#Age Distribution of the Participants
pt.figure(figsize=(10,5))
pt.title('Age distribution of the athletes')
pt.xlabel('Age')
pt.ylabel('Number of participant')
pt.hist(athletes.Age, bins=np.arange(10,80,2), color='orange', edgecolor='white')
pt.show()
```

```python
#Total number of female athletes in each olympic
female = athletes[(athletes['Sex']=='F') & (athletes['Season']=='Summer')][['Sex','Year']]
female = female.groupby('Year').count().reset_index()
```

```python
#Filtering female athletes in each olympic
woman = athletes[(athletes['Sex']=='F') & (athletes['Season']=='Summer')]
```

```python
pt.figure(figsize=(18,9))
sns.countplot(data = woman, x='Year')
pt.title('Woman participation')
```

```python
#line graph
plot = woman.groupby('Year')['Sex'].value_counts()
pt.figure(figsize=(15,7))
plot.loc[:,'F'].plot()
pt.title('Plot of Female Athletes over time')
```

```python
#Gold medal athletes
Gold = athletes[athletes['Medal'] =='Gold']
```

```python
Gold.head()
```

```python
#Take only the values that are different from NaN. 
Gold = Gold[np.isfinite(Gold['Age'])]
```

```python
#Gold beyond 60
Gold['ID'][Gold['Age']>60].count()
```

```python
#Checking sport where the 6 Gold medal has come
sporting = Gold['Sport'][Gold['Age']>60]
```

```python
sporting
```

```python
#Gold medals for each City
Gold['City'].value_counts().reset_index(name='Medal').head(5)
```

```python
TotalGoldMedals = Gold['City'].value_counts().reset_index(name='Medal').head(6)
ax = sns.catplot(x="index", y='Medal', data=TotalGoldMedals, kind='bar', height=5,)
ax.despine(left=True)
ax.set_xlabels('Top 5 city')
ax.set_ylabels('Number of Medals')
pt.title('Gold Medal per city')
```

```python
#rio olympics
max_year = athletes.Year.max()
print(max_year)

team_names = athletes[(athletes['Year']==max_year) & (athletes['Medal']=='Gold')].Team
team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)
pt.xlabel(None)
pt.ylabel('Countrywise Medals for the year 2016')
```

```python
not_null_medals = athletes[(athletes['Height'].notnull()) & (athletes['Weight'].notnull())]
```

```python
pt.figure(figsize=(20,10))
sns.scatterplot(data = not_null_medals, x='Height', y='Weight', hue='Sex')
pt.title('Height vs Weight of olympic Medalists')
```
